Suez to Bombay, India	Steamer (the Mongolia) across the Red Sea and the Indian Ocean.	13 days

Hong Kong to Yokohama, Japan	Steamer (the Carnatic) across the South China Sea, East China Sea, and the Pacific Ocean.	6 days

Yokohama to San Francisco, United States	Steamer (the General Grant) across the Pacific Ocean.	22 days

New York to London	Steamer (the China) across the Atlantic Ocean to Liverpool and rail.	9 days


Source: https://en.wikipedia.org/wiki/Around_the_World_in_Eighty_Days

```python
import geocoder

route = ['London', 'Suez', 'Bombay', 'Calcutta', 'Hong Kong',
         'Yokohama', 'San Francisco', 'New York City', 'London']
coords = []
for place in route:
    result = geocoder.osm(place)
    print('{} (long: {}, lat: {})'.format(place, result.x, result.y))
    coords.append([result.x, result.y])
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

xs, ys = zip(*coords)

ax = plt.axes(projection=ccrs.PlateCarree())

import shapely.geometry as sgeom


line = sgeom.LineString(coords)

# TODO: Investigate why plot is making polgons! :(
# ax.plot(xs, ys, transform=ccrs.Geodetic(), lw=5)

# TODO: Investigate why add_geometries is adding a facecolor. :(
ax.add_geometries([line], ccrs.Geodetic(), lw=5, facecolor='none', edgecolor='blue')

ax.coastlines()

ax.set_global()
```
