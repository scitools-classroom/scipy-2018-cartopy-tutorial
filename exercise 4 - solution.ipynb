```python
route = [('London', -0.1276474, 51.5073219),
         ('Suez', 32.537086, 29.974498),
         ('Bombay', 72.8882172, 19.1334321),
         ('Calcutta', 88.3476023, 22.5677459),
         ('Hong Kong', 114.1628131, 22.2793278),
         ('Yokohama', 139.636768, 35.444991),
         ('San Francisco', -122.4629897, 37.7647993),
         ('New York City', -73.9866136, 40.7306458),
         ('London', -0.1276474, 51.5073219)]
```

```python
import cartopy.crs as ccrs

import matplotlib.pyplot as plt
```

```python
%matplotlib notebook

plt.figure(figsize=(10, 10))

ax1 = plt.subplot(2, 1, 1, projection=ccrs.Mollweide())
name, xs, ys = zip(*route)
ax1.set_title('Shortest route between key waypoints')
ax1.plot(xs, ys, transform=ccrs.Geodetic(), linewidth=4, alpha=0.5)
ax1.set_global()
ax1.coastlines()
ax1.gridlines()

prj = ccrs.Robinson(central_longitude=11.25)
# These magic numbers come from geo-referencing the image.
# Take a look at the appendinx script image_geolocator.py
# for more info.
extent = [-13699119, 17123551, -6356356, 8605867]

ax2 = plt.subplot(2, 1, 2, projection=prj)
fname = 'resources/640px-Around_the_World_in_Eighty_Days_map.png'
img = plt.imread(fname)
ax2.set_title('Actual route (source wikipedia)')
ax2.imshow(img, extent=extent, transform=prj, origin='upper')
ax2.set_global()
ax2.coastlines()
ax2.gridlines()
```
